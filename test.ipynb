```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('bom.csv')  # Skip the first row as it's empty

# Function to normalize screw names (remove 'Mirror' suffix and normalize spacing)
def normalize_screw_name(name):
    if pd.isna(name):
        return name
    name = str(name)
    name = name.replace('(Mirror)', '').strip()
    return name

# Filter for rows containing screw-related terms and normalize names
screw_mask = df['Part Name'].str.startswith('M') & df['Part Name'].str.contains('M[0-9]', case=False, na=False)
screws_df = df[screw_mask].copy()
screws_df['Part Name'] = screws_df['Part Name'].apply(normalize_screw_name)

# Group by normalized part name and sum quantities
screw_summary = screws_df.groupby('Part Name')['Quantity'].sum().sort_values(ascending=False)

screw_summary
```
